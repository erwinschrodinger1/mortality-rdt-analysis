```python
import geopandas as gp
import pandas as pd
import matplotlib.pyplot as plt
from libpysal.weights.contiguity import Queen
from esda.moran import Moran_Local
from splot.esda import lisa_cluster
```

```python
country = gp.read_file("./local_unit/local_unit.shp")
country = country[['DISTRICT','Province','geometry']]
country = country.dissolve(by='DISTRICT')
```

```python
country
```

```python
country.plot()
```

## Filtering mortality data
### Steps Taken:
1. Matching the district name with shape file by upper casing district names
2. Changing names of district as of shape file
   - Parasi -> Nawalparasi_w
   - Nawalpur -> Nawalparasi_e
   - Rukum east -> Rukum_e
   - Rukum west -> Rukum_w

```python
mortality_data = pd.read_excel("./RTA-data Students.xlsx")
mortality_data["DISTRICT"] = mortality_data["DISTRICT"].str.upper()
mortality_data.loc[mortality_data[mortality_data['DISTRICT'] == 'PARASI'].index[0],"DISTRICT"] ="NAWALPARASI_W"
mortality_data.loc[mortality_data[mortality_data['DISTRICT'] == 'NAWALPUR'].index[0],"DISTRICT"] ="NAWALPARASI_E"
mortality_data.loc[mortality_data[mortality_data['DISTRICT'] == 'RUKUM EAST'].index[0],"DISTRICT"] ="RUKUM_E"
mortality_data.loc[mortality_data[mortality_data['DISTRICT'] == 'RUKUM WEST'].index[0],"DISTRICT"] ="RUKUM_W"
mortality_data
```

## Merging data of shp with mortality

```python
merged_mortality = country.merge(mortality_data,left_on ="DISTRICT",right_on="DISTRICT")
```

```python
merged_mortality
```

```python
merged_mortality = merged_mortality.sort_values(by="Total Death 2076_077").reset_index()
```

```python
merged_mortality
```

## Using quartile distribution using k=10 ie. Decile distribution to plot the data

```python
merged_mortality.plot(column="Total Death 2076_077",
                          legend=True,
                          cmap='GnBu_r',
                          scheme='quantiles',
                          k=10,
                          figsize=(15, 10),
                          edgecolor="white",
                          missing_kwds={
                                "color": "lightgrey",
                                "edgecolor": "red",
                                "hatch": "///",
                                "label": "Missing values"
                          })
```

## Calculating mortality per 100k people

```python
merged_mortality["PER_100K"] = (merged_mortality["Total Death 2076_077"]/merged_mortality["Population_CBS 2023"])*100000
```

```python
merged_mortality
```

## Constructing LISP map for further analysis

```python
y = merged_mortality['PER_100K'].values

w = Queen.from_dataframe(merged_mortality)

w.transform = 'r'

moran_loc = Moran_Local(y, w)
```

```python
fig = lisa_cluster(moran_loc, merged_mortality)
plt.show()
```
